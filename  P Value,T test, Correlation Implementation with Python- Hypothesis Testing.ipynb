## T test

A t-test is a type of inferential statistics which is used to determine if there is a sifnificant difference between the means if two groups which may be related in certain features

T-test has 2 types : 1 one sampled t-test 2.two-sampled t-test

## One sample T-test with python
The test will tell us whether means of the sample and the population are different based on the statistical calculation

```python
#population data
ages = [10,20,35,50,28,40,55,18,16,55,30,25,43,18,30,28,14,24,32,26,27,65,18,43,23,21,20,19,70]
```

### find it whether is there any difference between population mean and the sample mean

### H0-There is no difference
### H1-There is a differnce
### CI -95%
### significance -5% or 0.05%

```python
len(ages)
```

```python
import numpy as np
ages_mean = np.mean(ages)
print(ages_mean)
```

```python
#let the sample
sample_size = 10
age_sample = np.random.choice(ages,sample_size)
```

```python
#sample data
age_sample
```

```python
age_sample.mean()
```

```python
#apply this one numerical in t test
from scipy.stats import ttest_1samp
```

```python
#for any test implementation we will get 2 value one is particular test value based on the calculation another one is p value based on the final test value
ttest,p_value = ttest_1samp(age_sample,ages_mean) #based in the population mean ony we can able to do the t test and get the result based on the ho and h1
```

```python
p_value
```

```python
if p_value<0.05:
    print('We are rejecting the null hypothesis')
else:
    print('We are failed to reject the null hypothesis and accepting the alternative hypothesis')
```
